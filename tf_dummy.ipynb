```python
import os
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# define model
model = Sequential()
model.add(Dense(37, activation='tanh', kernel_initializer='he_normal', input_shape=(8,)))
tf.keras.layers.Dropout(0.5)
model.add(Dense(29, activation='tanh', kernel_initializer='he_normal'))
tf.keras.layers.Dropout(0.2)
model.add(Dense(1, activation='relu'))
```

```python
model.summary()
```

**Load data and begin tarining and testing**

```python
# training data
data= pd.read_csv(os.path.join('train_data - Sheet1.csv'))
d = data.to_numpy()
y = d[:,0]
x = d[:,1:9]
```

```python
# testing data
data2 = pd.read_csv(os.path.join('test_data - Sheet1.csv'))
d1 = data2.to_numpy()
y1 = d1[:,0]
x1 = d1[:,1:9]
```

*Dataset distribution*

```python
model.compile(optimizer='adam', loss = 'MeanSquaredError', metrics = ['accuracy'])
```

**trial 1**

```python
lr_schedule = keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=1e-1,
    decay_steps=10000,
    decay_rate=0.9)
optimizer = keras.optimizers.Adam(learning_rate=lr_schedule)
```

```python
model.compile(optimizer,
              loss=tf.keras.losses.LogCosh(),
              metrics=['accuracy'])
```

```python
history = model.fit(x, y, epochs = 10000)
```
